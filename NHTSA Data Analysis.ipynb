C24 Work Zone

Definition: This data element identifies a motor vehicle traffic crash in which the first harmful 
event occurs within the boundaries of a work zone or on an approach to or exit from a work 
zone, resulting from an activity, behavior, or control related to the movement of the traffic units 
through the work zone.

Additional Information: This data element identifies a â€œWork Zone Accidentâ€ as defined in 
ANSI D16.1, 7th Edition. If the crash qualifies as a â€œWork Zone Accidentâ€ then the type of work 
activity is identified. Use of the codes does not imply that the crash was caused by the 
construction, maintenance, or work activity.

The data element name was â€œConstruction/Maintenance Zoneâ€ from 1975 to 2008. The data

```python
# drop additional columns based on reviewing the column documentation in the NHTSA FARS Manual.

accident_drop_columns_list2 = ['RD_OWNER', 'VE_TOTAL', 'FUNC_SYS', 'LAND_USE',
                         'ROAD_FNC','FED_AID']

accidents_dropped = accidents_dropped.drop(accident_drop_columns_list2, 1)
print(accidents_dropped.shape)
```

```python
# values 0, 8, 9, 98, 99, and 99999 are all used for unobserved entries.
# using these values per documentation of columns to fill NA values.

fill_NA_0 = ['DRUNK_DR', 'PEDS', 'PVH_INVL', 'PERNOTMVIT', 'PERMVIT']
fill_NA_8 = ['SCH_BUS']
fill_NA_9 = ['NO_LANES', 'PROFILE', 'PAVE_TYP', 'SUR_COND', 'HIT_RUN',
             'T_CONT_F', 'NHS']
fill_NA_98 = ['TYP_INT']
fill_NA_99 = ['SP_LIMIT', 'TRA_CONT', 'CF1', 'CF2', 'CF3', 'HOSP_HR',
             'HOSP_MN']
fill_NA_99999 = ['MILEPT']
```

```python
#use lists above to fill missing values.

for column in fill_NA_0:
        accidents_dropped[column].fillna(0, inplace = True)
for column in fill_NA_8:
        accidents_dropped[column].fillna(8, inplace = True)
for column in fill_NA_9:
        accidents_dropped[column].fillna(9, inplace = True)
for column in fill_NA_98:
        accidents_dropped[column].fillna(98, inplace = True)
for column in fill_NA_99:
        accidents_dropped[column].fillna(99, inplace = True)
for column in fill_NA_99999:
        accidents_dropped[column].fillna(99999, inplace = True)
        
accidents_dropped.isnull().sum()
```

```python
# convert FLOAT types to INT types.  
# The FARS manual uses whole numbers for these columns.

to_int_type = ['NO_LANES', 'SP_LIMIT', 'PROFILE', 'PAVE_TYP', 'SUR_COND',
              'TRA_CONT', 'HIT_RUN', 'SCH_BUS', 'CF1', 'CF2', 'CF3',
              'DRUNK_DR', 'T_CONT_F', 'HOSP_HR', 'HOSP_MN', 'PEDS', 'NHS',
              'TYP_INT', 'PVH_INVL', 'PERNOTMVIT', 'PERMVIT', 'TRAFFIC_FLOW',
              'ROAD_TYPE', 'WORK_ZONE', 'REL_JCT']

for column in to_int_type:
    accidents_dropped[column] = accidents_dropped[column].astype(int)

accidents_dropped.dtypes
```

```python
# reduce the size of the dataframe by reducing to int32 for all columns.

for column in accidents_dropped.columns:
    if accidents_dropped[column].dtype == 'int64':
        accidents_dropped[column] = accidents_dropped[column].astype('int32')

accidents_dropped.dtypes
```

```python
# 99 = Unknown and 24 was used from 1975-2008. Distribute the 99's across the other hours.
# Change 24 to 23 as corrected from 2009-2021.

from random import randrange

for hour in df_a['HOUR'].values:
    if hour == 99:
        df_a['HOUR'].replace(hour, randrange(0, 24), inplace=True)             
    if hour == 24:
        df_a['HOUR'].replace(hour, 23, inplace=True) 

df_a['HOUR'].value_counts()
```
